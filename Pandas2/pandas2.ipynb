```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Problem 1

```python
def prob1():
    """
    Create 3 visualizations of the crime data set.
    One of the visualizations should be a histogram.
    The visualizations should be clearly labelled and easy to understand.
    """    
    
    #read in data
    data = pd.read_csv('crime_data.csv', delimiter=',', index_col='Year')
    #plot violent crimes
    data.plot(y=['Murder','Forcible Rape','Robbery','Aggravated Assault'], subplots=True, layout=(2, 2), sharey=False, style=['-', '--', '-.', ':'], title='Violent Crimes')
    #plot non violent crimes
    data.plot(y=['Property', 'Burglary', 'Larceny','Vehicle Theft'], subplots=True, layout=(2, 2), sharey=False, style=['-', '--', '-.', ':'], title='NonViolent Crimes')
    
    #plot population and total crimes on a histogram
    data[['Population', 'Total']].hist(grid=False, bins=10)
    
    return 
```

```python
prob1()
```

# Problem 2

```python
 def prob2():
    """
    Use a visualization for each of the following to identify if trends exists 
    between Forcible Rape and
        1. Violent
        2. Burglary
        3. Aggravated Assault
    Plot each visualization.
    Return a tuple of booleans. Each entry should identify whether that 
    element correlates or not.
    """
    #read in data
    data = pd.read_csv('crime_data.csv', delimiter=',', index_col='Year')
    #drop un-needed data
    data.drop(['Population', 'Total', 'Property', 'Murder', 'Robbery', 'Larceny', 'Vehicle Theft'], inplace=True, axis=1)

    #creae scatter matrix plots to determine correlations
    pd.plotting.scatter_matrix(data[['Forcible Rape','Violent']])
    pd.plotting.scatter_matrix(data[['Forcible Rape','Burglary']])
    pd.plotting.scatter_matrix(data[['Forcible Rape','Aggravated Assault']])

    return True, False, True
```

```python
print('Are there trends between Forcible Rape and:', ('Violent', 'Burglary', 'Aggravted Assault'), prob2())
```

# Problem 3

```python
def prob3():
    """
    Use crime_data.csv to display the following distributions.
        1. The distributions of Burglary, Violent, and Vehicle Theft
        2. The distributions of Vehicle Theft against the values of Robbery
    """
    #read in data
    data = pd.read_csv('crime_data.csv', delimiter=',', index_col='Year')
    #drop unwanted data
    data.drop(['Population', 'Total', 'Property', 'Murder', 'Larceny','Forcible Rape', 'Aggravated Assault', 'Larceny'], inplace=True, axis=1)
    
    #plot distrbiutions of Burglary, Volent crimes, and Vehicle Theft
    data.drop(['Robbery'], axis=1).plot(kind='box', vert=False, title='Distributions of Violent, Burglary, and Vehicle Theft')
    
    #drop unwanted data
    data.drop(['Violent', 'Burglary'], axis=1, inplace=True)
    #create a hexbin to plot vehicle theft data against robbery data
    data.plot(kind='hexbin', x='Vehicle Theft', y='Robbery', sharex=False, gridsize=15, title='Vehicl Theft vs Robbery')

    return
```

```python
prob3()
```
