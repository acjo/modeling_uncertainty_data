```python
import numpy as np
from scipy.stats import norm
from matplotlib import pyplot as plt
```

```python
def plot_pdf_cdf(mu, sigma):
    x = np.linspace(norm.ppf(0.001, loc=mu, scale=np.sqrt(sigma)), norm.ppf(0.999, loc=mu, scale=np.sqrt(sigma)), 1000)
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.plot(x, norm.pdf(x), 'r-', lw=2, alpha=1, label='Normal pdf')
    ax.legend(loc='best')
    y = np.linspace(norm.ppf(0.001, loc=mu, scale=np.sqrt(sigma)), norm.ppf(0.999, loc=mu, scale=np.sqrt(sigma)), 1000)
    ax.plot(y, norm.cdf(y), 'b-', lw=2, alpha=1, label='Normal cdf')
    ax.legend(loc='best')
    plt.suptitle('Normal distribution with ' + r'$\mu = $' + str(mu) + ' and '+ r' $\sigma^2 = $' + str(sigma))
    plt.show()
                             

```

```python
plot_pdf_cdf(0, 1)
```

```python
plot_pdf_cdf(0, 5)
```
