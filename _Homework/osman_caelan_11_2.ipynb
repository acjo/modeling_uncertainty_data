```python
#########################################
# CODE TO CALCULATE PROBS
#########################################
def P(S=[], G=[]):
    As = ['A', 'nA']
    Bs = ['B', 'nB']
    Es = ['E', 'nE']
    Js = ['J', 'nJ']
    Ms = ['M', 'nM']
    if len(G)==0:
        if len(S) == 1:
            if S[0] == 'B':
                return 0.002
            elif S[0] == 'nB':
                return 1 - P(S=['B'])
            elif S[0] == 'E':
                return 0.001
            elif S[0] == 'nE':
                return 1 - P(S=['E'])
            elif S[0] == 'J':
                PJ_1 = 0
                for A in As:
                    for B in Bs:
                        for E in Es:
                            for M in Ms:
                                PJ_1 += P(S=[B])*P(S=[E])*P(S=[A], G=[B, E])\
                                *P(S=[M], G=[A])*P(S=['J'], G=[A])
                return PJ_1
            elif S[0] == 'A':
                PA_1 = 0
                for B in Bs:
                    for E in Es:
                        for J in Js:
                            for M in Ms:
                                PA_1 += P(S=[B])*P(S=[E])*P(S=['A'], G=[B, E])\
                                *P(S=[M], G=['A'])*P(S=[J], G=['A'])
                return PA_1
    elif len(G) == 1:
        if len(S) == 1:
            if S[0] == 'J' and G[0] == 'A':
                return 0.9
            elif S[0] == 'nJ' and G[0] == 'A':
                return (1 - P(S=['J'], G=['A']))
            elif S[0] == 'M' and G[0] == 'A':
                return 0.7
            elif S[0] == 'nM' and G[0] == 'A':
                return (1 - P(S=['M'], G=['A']))
            elif S[0] == 'J' and G[0] == 'nA':
                return 0.05
            elif S[0] == 'nJ' and G[0] == 'nA':
                return (1 - P(S=['J'], G=['nA']))
            elif S[0] == 'M' and G[0] == 'nA':
                return 0.01
            elif S[0] == 'nM' and G[0] == 'nA':
                return (1 - P(S=['M'], G=['nA']))
            elif S[0] == 'J' and G[0] == 'B':
                num = 0
                for A in As:
                    for E in Es:
                        for M in Ms:
                            num += P(S=['B'])*P(S=[E])\
                            *P(S=[A], G=['B', E])*P(S=[M], G=[A])*P(S=['J'], G=[A])
                return num/P(S=['B'])
            elif S[0] == 'E' and G[0] == 'A':
                num = 0
                for B in Bs:
                    for J in Js:
                        for M in Ms:
                            num += P(S=[B])*P(S=['E'])*P(S=['A'], G=[B, 'E'])\
                            *P(S=[J], G=['A'])*P(S=[M], G=['A'])
                return num/P(S=['A'])
            elif S[0] == 'B' and G[0] == 'A':
                num = 0
                for E in Es:
                    for J in Js:
                        for M in Ms:
                            num += P(S=['B'])*P(S=[E])*P(S=['A'], G=['B', E])\
                            *P(S=[J], G=['A'])*P(S=[M], G=['A'])
                return num/P(S=['A'])
        if len(S) == 2:
            if S[0] == 'B' and S[1] == 'E' and G[0] == 'A':
                num = 0
                for J in Js:
                    for M in Ms:
                        num += P(S=['B'])* P(S=['E'])*P(S=['A'], G=['B', 'E'])\
                        *P(S=[J], G=['A'])*P(S=[M], G=['A'])
                return num/P(S=['A'])
    elif len(G) == 2:
        if len(S) == 1:
            if S[0] == 'A' and G[0]== 'B' and G[1]=='E':
                return 0.95
            elif S[0] == 'A' and G[0] == 'B' and G[1] == 'nE':
                return 0.94
            elif S[0] == 'A' and G[0] == 'nB' and G[1] == 'E':
                return 0.29
            elif S[0] == 'A' and G[0] == 'nB' and G[1] == 'nE':
                return 0.001
            elif S[0] == 'nA' and G[0]== 'B' and G[1]=='E':
                return 1-0.95
            elif S[0] == 'nA' and G[0] == 'B' and G[1] == 'nE':
                return 1-0.94
            elif S[0] == 'nA' and G[0] == 'nB' and G[1] == 'E':
                return 1-0.29
            elif S[0] == 'nA' and G[0] == 'nB' and G[1] == 'nE':
                return 1-0.001
            elif S[0] == 'B' and G[0] == 'J' and G[1] == 'nM':
                num = 0
                denom = 0
                for A in As:
                    for E in Es:
                        num += P(S=['B'])*P(S=[E])*P(S=[A], G=['B', E])\
                        *P(S=['nM'], G=[A])*P(S=['J'], G=[A])
                        for B in Bs:
                            denom +=P(S=[B])*P(S=[E])*P(S=[A], G=[B, E])\
                            *P(S=['nM'], G=[A])*P(S=['J'], G=[A])
                return num/denom
            
print('P(J=1):',P(S =['J']))
print('P(J=1|B=1):', P(S=['J'], G=['B']))
print('P(B=1|J=1, M=0):', P(S=['B'], G=['J', 'nM']))
print('P(E=1 |A=1):', P(S=['E'], G=['A']))
print('P(B=1 |A=1):', P(S=['B'], G=['A']))
print('P(B=1, E=1 |A=1):', P(S = ['B', 'E'], G=['A']))
print('P(E=1 |A=1)*P(B=1 |A=1):', P(S=['E'], G=['A'])*P(S=['B'], G=['A']))
```

(ii) 

\begin{align}
f_{(X, Y)|Z\geq 0} (x, y) &= f_{(X, Y)|X+Y \geq 0}(x, y)
\end{align}

Notice this means that we will only accept a draw from the joint p.d.f. of $(X, Y)$ if $X+Y \geq 0$. 
Therefore this is the same as drawing first and and then rejecting the draw based on if the sum is positive or negative. 
Therefore
\begin{align}
f_{(X, Y)|Z\geq 0}(x, y,z) \propto f_{(X, Y)}(x, y) \mathbb{1}_{(X, Y)\geq 0}(x,  y)
\end{align}


```python
import numpy as np
from scipy.stats import multivariate_normal
from matplotlib import pyplot as plt

#part iii
#draw
draw = multivariate_normal.rvs(mean = np.zeros(2), cov = np.eye(2), size=int(1e5))
#plot
fig = plt.figure()
fig.set_dpi(150)
ax = fig.add_subplot(111)
ax.plot(draw[:, 0], draw[:, 1], 'r,')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Multivariate Draw')
plt.show()

#part 4
#Get Z draw
sums = draw.sum(axis=1)
#get conditioned on Z draw
conditioned_draw = draw[sums>=0]

fig = plt.figure()
fig.set_dpi(150)
ax = fig.add_subplot(111)
ax.plot(conditioned_draw[:, 0], conditioned_draw[:, 1], 'b,')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title(r'Draw from $Z$')
plt.show()

#part 5
sample_cov = np.cov(conditioned_draw.T)
print('covariance between X and Y:', sample_cov[-1, 0])

#part 6
E_X_Z = np.mean(conditioned_draw[:, 0])
E_Y_Z = np.mean(conditioned_draw[:, 1])

print('E[X|Z >= 0]:', E_X_Z)
print('E[Y|Z >= 0]:', E_Y_Z)
```
