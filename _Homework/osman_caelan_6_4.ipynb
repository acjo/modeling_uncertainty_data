# Exercise 6.17

```python
def problem6_17():
    BC = datasets.load_breast_cancer()
    X_bc = BC.data
    y_bc = BC.target

    DIG = datasets.load_digits()
    X_dig = DIG.data
    y_dig = DIG.target == 4

    IRIS = datasets.load_iris()
    X_ir = IRIS.data
    y_ir = IRIS.target == 0

    trainers = {'Breat Cancer': [X_bc, y_bc], 'Digits': [X_dig, y_dig], "Iris" : [X_ir, y_ir]}
    classifiers = ['Decision Tree Classifier', 'Logisitc Regression', 'Gaussian Naive Bayes']

    for class_  in classifiers:
        for trainer in list(trainers.keys()):
            X_train, X_test, y_train, y_test = train_test_split(trainers[trainer][0],
                                                                trainers[trainer][-1], 
                                                                train_size = 0.7)
            if class_ == 'DecisionTreeClassifier':
                RFC = DecisionTreeClassifier().fit(X_train, y_train)
                RocCurveDisplay.from_estimator(RFC, X_test, y_test)
                plt.title(class_ + ' ' + trainer)
                plt.show()
            elif class_ == 'LogisticRegression':
                LR = LogisticRegression().fit(X_train, y_train)
                RocCurveDisplay.from_estimator(LR, X_test, y_test)
                plt.title(class_ + ' ' + trainer)
                plt.show()
            else:
                GNB = GaussianNB().fit(X_train, y_train)
                RocCurveDisplay.from_estimator(GNB, X_test, y_test)
                plt.title(class_ + ' ' + trainer)
                plt.show()

    return

problem6_17()
```
